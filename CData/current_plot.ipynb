```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv('cdata_3oct_simul.csv')

# Extract x and y
x = df['TimeStamp_N']  # Ensure numerical or preprocess if necessary
y = df['Value_A']      # Ensure numerical or preprocess if necessary

# Convert to numerical if needed
x = pd.to_numeric(x, errors='coerce')
y = pd.to_numeric(y, errors='coerce')

# Drop NaN values if conversion fails for some rows
df = df.dropna(subset=['TimeStamp_N', 'Value_A'])

# Recompute x and y after cleaning
x = df['TimeStamp_N']
y = df['Value_A']

# Calculate mean
mean = np.mean(y)

# Create the plot
plt.figure(figsize=(14, 6))

plt.title('Healthy condition (Current Vs Time)')
plt.ylabel('Current (A)', fontsize=15)
plt.xlabel('Time (s)', fontsize=15)

plt.plot(x, y, color='y', label='Normal')
plt.axhline(y=mean, color='y', linestyle='--', label=f'Mean: {mean:.2f}')

# Adjust xticks dynamically based on x range

xticks_range = np.arange(0,1022,105)
plt.xticks(xticks_range, rotation=20)

# Adjust y limits if needed
plt.ylim(7.5, 8.5)

plt.legend()
plt.show()
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('cdata.csv');

x = df['TimeStamp_C'];
y1 = df['Value_N'];
y2 = df['Value_30'];
y3 = df['Value_45'];
y4 = df['Value_60'];

mean1 = np.mean(y1);
mean2 = np.mean(y2);
mean3 = np.mean(y3);
mean4 = np.mean(y4);

plt.figure(figsize=(14,6));
plt.title('Variation of Current Vs Time');
plt.xlabel('Time(s)', fontsize=15);
plt.ylabel('Current (A)', fontsize=15);

plt.plot(x,y1, label='Normal Condition');
plt.plot(x,y2, label='Upto 30 deg Close');
plt.plot(x,y3, label='Upto 45 deg Close');
plt.plot(x,y4, label='Upto 60 deg Close');

plt.axhline(y=mean1, color='b', linestyle='--', label=f'Mean at Normal/Healthy : {mean1:.2f}')
plt.axhline(y=mean2, color='orange', linestyle='--', label=f'Mean Upto 30 deg close: {mean2:.2f}')
plt.axhline(y=mean3, color='g', linestyle='--', label=f'MeanUpto 45 deg close: {mean3:.2f}')
plt.axhline(y=mean4, color='r', linestyle='--', label=f'MeanUpto 60 deg close: {mean4:.2f}')

plt.xticks(np.arange(0,1150,100),rotation=20);
plt.xlim(1,1050)
plt.ylim(7.5, 10.2)

plt.legend();
plt.show();
```
