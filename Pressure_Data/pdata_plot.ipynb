```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('pdata_3oct_simul.csv');

x = df['TimeStamp_N'];
y = df['Values_N'];
plt.figure(figsize=(14,6));

mean = np.mean(y);

plt.title('Pressure Vs Time At Normal Condition')
plt.ylabel('Pressure(KPa)',fontsize=15);
plt.xlabel('Time(s)',fontsize=15);

plt.plot(x,y, label='Normal');

plt.axhline(y=mean, color='b', linestyle='--', label=f'Mean: {mean: .2f}')

plt.xticks(np.arange(0,1300,100))
plt.xlim(55,1300)

plt.legend()
plt.show();
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('pdata.csv');
# 
x = df['TimeStamp_C'];
y1 = df['Values_N'];
y2 = df['Values_30'];
y3 = df['Values_45'];
y4 = df['Values_60'];
plt.figure(figsize=(14,6));

mean1 = np.mean(y1);
mean2 = np.mean(y2);
mean3 = np.mean(y3);
mean4 = np.mean(y4);

plt.title('Variation of Pressure Vs Time')
plt.ylabel('Pressure(KPa)',fontsize=15);
plt.xlabel('Time(s)',fontsize=15);

plt.plot(x,y1, label='Normal Condition');
plt.plot(x,y2, label='Upto 30 deg close');
plt.plot(x,y3, label='Upto 45 deg close');
plt.plot(x,y4, label='Upto 60 deg close');

plt.axhline(y=mean1, color='b', linestyle='--', label=f'Mean At Normal: {mean1: .2f}');
plt.axhline(y=mean2, color='y', linestyle='--', label=f'Mean Upto 30 deg: {mean2: .2f}');
plt.axhline(y=mean3, color='g', linestyle='--', label=f'Mean Upto 45 deg: {mean3: .2f}');
plt.axhline(y=mean4, color='r', linestyle='--', label=f'Mean Upto 60 deg: {mean4: .2f}');

plt.xticks(np.arange(0,1348,100));
plt.xlim(0,1300);
plt.ylim(5,110);
#
plt.legend();
plt.show();
```
