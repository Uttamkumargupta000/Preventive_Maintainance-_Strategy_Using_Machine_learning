```python
 
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
# importing the models
 
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier  # Import XGBoost
```

```python
y = df['fault_type']  # Assign the 'fault_type' column to y
x = df.iloc[:, :-1]   # Exclude the last column and assign the remaining DataFrame to x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.20, random_state=42)
```

```python
 # List of models
models_list = [
    
    SVC(),
    KNeighborsClassifier(),
    RandomForestClassifier(random_state=0),
    XGBClassifier(use_label_encoder=False, eval_metric='logloss')  # Add XGBoost
]
```

```python
model_hyperparameters = {
    'svc_hyperparameters': {
        'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
        'C': [1, 5, 10, 20],
         
    },

    'KNN_hyperparameters': {
        'n_neighbors': [3, 5, 10, 20],
        
        'p': [1, 2],  # Distance metric: 1=Manhattan, 2=Euclidean
   
    },

    'random_forest_hyperparameters': {
        'n_estimators': [10, 20, 50, 100, 200],  # Number of trees
        'max_depth': [None, 10, 20, 30],  # Depth of the tree
        
        'criterion': ['gini', 'entropy']  # Function to measure the quality of a split
    },

    'xgboost_hyperparameters': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 0.2, 0.3],  # Step size shrinkage
        'max_depth': [3, 6, 10],  # Maximum depth of a tree
        
    }
}
```

```python
def ModelSelection(list_of_models, hyperparameters_dictionary):

  result = []

  i = 0

  for model in list_of_models:

    key = model_keys[i]

    params = hyperparameters_dictionary[key]

    i += 1

    print(model)
    print(params)
    print('---------------------------------')


    classifier = GridSearchCV(model, params, cv=5)

    # fitting the data to classifier
    # x_scaled = scaler.fit_transform(x)
    classifier.fit(X_train,y_train)

    result.append({
        'model used' : model,
        'highest score' : classifier.best_score_,
        'best hyperparameters' : classifier.best_params_
    })

  result_dataframe = pd.DataFrame(result, columns = ['model used','highest score','best hyperparameters'])

  return result_dataframe
```

```python
model_keys = list(model_hyperparameters.keys())
print(model_keys)
```

```python
results = ModelSelection(models_list, model_hyperparameters )
print(results)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=0,criterion='entropy',max_depth=None,n_estimators=10)

clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(16, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

average (optional): This parameter defines the type of averaging performed on the data. It can take several values:

    'micro': Calculates metrics globally by counting the total true positives, false negatives, and false positives.
    'macro': Calculates metrics for each label and finds their unweighted mean. This does not take label imbalance into account.
    'weighted': Calculates metrics for each label and finds their average, weighted by support (the number of true instances for each label).
    None: Returns the score for each class.

```python
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
```
